```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)

# Initialize Sentiment Intensity Analyzer
sid = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def analyze_sentiment(text):
    print("\nSentence: ", text)
    sentence_scores = sid.polarity_scores(text)
    print("Sentence Sentiment Scores:")
    for key in sorted(sentence_scores):
        print('{0}: {1}, '.format(key, sentence_scores[key]), end='')
    print("\n")

    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words and word not in punctuations]
    print("Word Sentiment Scores: ")
    for word in filtered_words:
        word_scores = sid.polarity_scores(word)
        print("Word: ", word)
        for key in sorted(word_scores):
            print('  {0}: {1}, '.format(key, word_scores[key]), end='')
        print("\n")
    print('\n')
```

```python
# Read the text file and analyze sentiment
with open('/content/drive/MyDrive/Colab Notebooks/Sentimental_Analysis/kindle.txt', encoding='ISO-8859-2') as f:
    for line in f.read().split('\n'):
        analyze_sentiment(line)
```

```python
# Get input from the user
user_input = input("Enter a sentence to analyze its polarity: ")
analyze_sentiment(user_input)
```
